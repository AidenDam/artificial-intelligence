# Uczenie zespo≈Çowe - Las losowy

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Generowanie zestawu danych

```
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
```

### Drzewo decyzyjne

```
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=10)
clf.fit(X, y)

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, y, clf)
```

```
clf.score(X, y)
```

### Metody zespo≈Çowe
#### Bagging Classifier

```
from sklearn.ensemble import BaggingClassifier

tree = DecisionTreeClassifier()
bag = BaggingClassifier(base_estimator=tree, n_estimators=100, max_samples=0.8)
bag.fit(X, y)
```

```
plot_decision_regions(X, y, bag)
```

```
bag.score(X, y)
```

#### Random Forest

```
from sklearn.ensemble import RandomForestClassifier

# uczenie zespo≈Çowe dla 100 drzew decyzyjnych (n_estimators=100)
# predykcje sƒÖ tworzone na podstawie losowania w≈õr√≥d tych modeli
forest = RandomForestClassifier(n_estimators=100)
forest.fit(X, y)
plot_decision_regions(X, y, forest)
```

```
forest.score(X, y)
```
